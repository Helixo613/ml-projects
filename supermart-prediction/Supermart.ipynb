# Date Features Extraction
Extracting additional features from the order and ship dates.

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Ship Delay'] = (df['Ship Date'] - df['Order Date']).dt.days
df[['Order Date', 'Ship Date', 'Order Year', 'Order Month', 'Ship Delay']].head()
```

```python
df.head()
```

# Data Encoding and cleaning
Encoding categorical features into numerical values.

```python
#Encoding data
df_cleaned = df.drop(['Order ID', 'Customer ID', 'Customer Name', 'Product ID', 'Product Name'], axis=1)
df_encoded = pd.get_dummies(df_cleaned, drop_first=True)
df_encoded.head()
```

```python
df_encoded['Profitable']=df_encoded['Profit'].apply(lambda x: 1 if x > 0 else 0)
```

```python
#Removing outliers
from scipy import stats
numeric_data=df_encoded.select_dtypes(include=[np.number])
z_score=np.abs(stats.zscore(numeric_data))
df_no_outliers = df_encoded[(z_score < 3).all(axis=1)]
```

```python
df_no_outliers.head()
```

```python
df_no_outliers.shape
```

```python
#Applying Scaling
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df_scaled=df_no_outliers.select_dtypes(include=[np.number])
df_scaled.head()
```

# Feature and Target Separation
Separating the target variable from the input features.

```python
from sklearn.model_selection import train_test_split
X=df_scaled.drop(['Profit'],axis=1)
Y=df_scaled['Profit']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
X_class = df_scaled.drop(['Profit'], axis=1)
y_class = df_encoded.loc[X_class.index, 'Profitable']
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_class, y_class, test_size=0.2, random_state=42)
```

# Model Training and Prediction
Training a regression model and making predictions.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

model=LinearRegression()
model.fit(X_train,Y_train)
Y_pred= model.predict(X_test)
r2 = r2_score(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
print(f'RÂ² Score: {r2}')
print(f'Mean Squared Error: {mse}')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(Y_test, Y_pred,alpha=0.5)
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'r--')
plt.xlabel('Actual Profit')
plt.ylabel('Predicted Profit')
plt.title('Actual vs Predicted Profit')
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

log_model = LogisticRegression(class_weight='balanced', max_iter=1000, solver='liblinear')
log_model.fit(X_train_clf, y_train_clf)
y_pred_clf = log_model.predict(X_test_clf)

print("Accuracy:", accuracy_score(y_test_clf, y_pred_clf))
print("Confusion Matrix:\n", confusion_matrix(y_test_clf, y_pred_clf))
print("Classification Report:\n", classification_report(y_test_clf, y_pred_clf))
```
